## Orientation tuning from neurons in the lateral geniculate nucleus of mouse thalamus
We can examine the spikes, mean firing rates, and variance of firing rates with data recorded from the mouse visual thalamus. In the `orientation_tuning_data` folder, are the neural responses of two neurons to a number of different orientations of static gratings. Files are named with the pattern `orientation` + angle + `_cell` + cell number + `.npy`. These are numpy files that you can load into a numpy array with `np.load()`. The orientation corresponds to the orientation of a static grating that was displayed to a mouse for 250ms. The gratings are shown below: 

0: ![0](orientation_0.png) 
30: ![30](orientation_30.png) 
60: ![60](orientation_60.png) 
90: ![90](orientation_90.png) 
120: ![120](orientation_120.png) 
150: ![150](orientation_150.png)

The data in the folder are the firing rates in the 250ms after the static grating appears on the screen, binned into spike counts in 1ms bins. There are 47-49 trials per orientation. As with the previous exercise, compute the mean and standard deviation of the firing rate for cell1 and cell2 across all orientations. Generate a tuning curve for each cell. Plot the tuning curves of both cells in the same graph, with orientation on the horizontal axis and mean firing rate across trials on the vertical axis. Include error bars to show the standard deviation of the firing rates for eachWould you be able to predict what orientation was being shown to the mouse based on the firing rate of cell1 and cell2? Are there orientations you would not be able to distinguish? 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Solutions:
orientations = [0, 30, 60, 90, 120, 150]
fig = plt.figure()
ax = fig.add_subplot(111)

   
for cell in ["cell1", "cell2"]:
    means = []
    sds = []
    for orientation in orientations:
        cell_data = np.load("orientation_tuning_data/orientation" + str(orientation) + "_" + cell + ".npy")
        cell_fr = np.sum(cell_data, axis=1)/0.25
        mean_cell = np.mean(cell_fr)
        std_cell = np.std(cell_fr)
        means.append(mean_cell)
        sds.append(std_cell)
    ax.errorbar(orientations, means, yerr=sds, label=cell, marker='o')
ax.legend()
ax.set_xlabel("Orientation Angle")
ax.set_ylabel("Firing Rate (sp/s)")
plt.show()
```
